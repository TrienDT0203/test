```python
def unplayed_songs(music_collection,id):
    update_music_collection = music_collection.copy()
    for album in update_music_collection:
        if str(album.id) == str(id):
            unplayed_songs = []
            for song in album.song:
                if song.play_count == 0:
                    unplayed_songs.append(song)
            if len(unplayed_songs) > 0:
                return unplayed_songs
    
    print('Album not found')
    return music_collection
            
```

7. Write a function called favorite_album that takes a music collection and returns the album that is the "favorite." We'll dene the favorite album as the one that has the highest playing time. (An album's playing time is the sum of all its songs' playing time.)

```python
def favorite_album(music_collection):
    highest_playing_time_album = {'album':0, 
                                  'total_playing_time':0}
    for album in music_collection:
        total_playing_time = 0
        for song in album.song:
            total_playing_time += song.length * song.play_count
        if total_playing_time > highest_playing_time_album['total_playing_time']:
            highest_playing_time_album['album'] = album
            highest_playing_time_album['total_playing_time'] = total_playing_time
    return highest_playing_time_album['album']
            
```

8. Write a function called unplayed_albums that takes a music collection and returns a list of the albums that have not been played.

```python
def unplayed_albums(music_collection):
    not_play_time_album = {'album':0, 
                            'total_play_count':0}
    
    for album in music_collection:
        total_play_count = 0
        for song in album.song:
            total_play_count += song.play_count
            
        if total_play_count == not_play_time_album['total_play_count']:
            not_play_time_album['album'] = album
            not_play_time_album['total_play_count'] = total_play_count
    
    if not_play_time_album['album'] == 0:
        print('Album not found')
        return music_collection
        
    return not_play_time_album['album']
    
            
```
